```python
from PIL import *
img = Image.open("westbrook.jpg")
width, height =img.size
print(width, height)

for y in range(0,height):
    for x in range(0,width):
        rgba=img.getpixel((x,y))
        rgba=(int(rgba[0]/2),
              int(rgba[1]/2),
              int(rgba[2]/2))
              
        img.putpixel((x,y),rgba)
    
img.show()
img.save('機器學習Udemy prerequisites/Q2.jpg')
```
